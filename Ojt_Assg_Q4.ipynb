**Polynomial Plotting**

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
np.random.seed(14)
coefficients = np.random.randn(6)
print(coefficients)
```

```python
X=np.arange(-5,6.1,0.1)
plt.plot(X,1.55133911*X**5+0.07918602*X**4+0.17397653*X**3-0.07233657 *X**2-2.0043294*X+0.14467781)
plt.annotate('local maxima',xy=(6,12500),xytext=(0,12500),arrowprops=dict(facecolor='black',shrink=0.0))
plt.annotate('local minima',xy=(-5,-5000),xytext=(-3,-5000),arrowprops=dict(facecolor='black',shrink=0.0))
```
